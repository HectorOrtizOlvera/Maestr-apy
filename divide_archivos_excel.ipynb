```python
!pip install xlwt
import pandas as pd
import xlwt # Explicitly import xlwt

# 1. Coloca aquí el nombre exacto de tu archivo original
# Se ha corregido el nombre del archivo y la ruta para que coincida con el archivo .xls en sample_data
archivo_original = "DF55M.xls"

# 2. Leemos el archivo de Excel
df = pd.read_excel(archivo_original)

# 3. Definimos el tamaño de los bloques (20 registros)
registros_por_archivo = 20

# 4. Dividimos el archivo usando un bucle
for i in range(0, len(df), registros_por_archivo):
    # Cortamos el fragmento de 20 filas
    bloque = df.iloc[i : i + registros_por_archivo]

    # Creamos un nombre numerado para cada archivo nuevo con extensión .xls
    numero_archivo = (i // registros_por_archivo) + 1
    nombre_salida = f"archivo_dividido_{numero_archivo}.xls"

    # Guardamos el bloque en un nuevo archivo Excel usando xlwt directamente
    try:
        workbook = xlwt.Workbook()
        sheet = workbook.add_sheet('Sheet1') # Add a sheet

        # Write headers
        for col_idx, col_name in enumerate(bloque.columns):
            sheet.write(0, col_idx, col_name)

        # Write data rows
        for row_idx, row in enumerate(bloque.values):
            for col_idx, cell_value in enumerate(row):
                sheet.write(row_idx + 1, col_idx, cell_value)

        workbook.save(nombre_salida)
        print(f"Creado con éxito: {nombre_salida}")
    except Exception as e:
        print(f"Error al escribir {nombre_salida} con xlwt directamente: {e}")

print("¡Listo! Tu archivo se ha dividido en partes de 20 registros.")
# Re-ejecutado después de intentar varias soluciones para xlwt
```

```python
import os

# Lista para almacenar los nombres de los archivos eliminados
archivos_eliminados = []

# Recorre todos los archivos en el directorio actual
for file in os.listdir('.'):
    # Verifica si el archivo es uno de los archivos divididos y termina en .xls o .xlsx
    if file.startswith('archivo_dividido_') and (file.endswith('.xls') or file.endswith('.xlsx')):
        try:
            os.remove(file)
            archivos_eliminados.append(file)
        except OSError as e:
            print(f"Error al eliminar el archivo {file}: {e}")

if archivos_eliminados:
    print("Archivos temporales eliminados:")
    for f in archivos_eliminados:
        print(f"- {f}")
else:
    print("No se encontraron archivos temporales .xls o .xlsx para eliminar.")
```

```python
import os
import zipfile
from google.colab import files

# Nombre del archivo ZIP de salida para los archivos .xls
zip_filename_xls = 'archivos_divididos_xls.zip'

# Crear un archivo ZIP
with zipfile.ZipFile(zip_filename_xls, 'w') as zipf:
    for root, dirs, files_in_dir in os.walk('.'):
        for file in files_in_dir:
            # Busca los archivos generados que comienzan con 'archivo_dividido_' y terminan en '.xls'
            if file.startswith('archivo_dividido_') and file.endswith('.xls'):
                file_path = os.path.join(root, file)
                zipf.write(file_path, arcname=file) # Añadir el archivo al ZIP con su nombre original

print(f"Se ha creado el archivo ZIP: {zip_filename_xls}")

# Descargar el archivo ZIP
files.download(zip_filename_xls)
```
