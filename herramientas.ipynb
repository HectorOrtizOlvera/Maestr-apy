codigo para acortar ligas urls

```python
pythonimport requests

def acortar_url(url_larga):
    api = "http://tinyurl.com"
    respuesta = requests.get(api + url_larga)
    return respuesta.text

mi_url = "https://tu-sitio-web-muy-largo.com"
print("URL corta:", acortar_url(mi_url))
```

genera contraseñas

Crea contraseñas aleatorias combinando letras, números y caracteres especiales. Es una herramienta nativa, por lo que no requiere instalar nada adicional.

```python
import string
import secrets

def generar_contrasena(longitud=12):
    caracteres = string.ascii_letters + string.digits + string.punctuation
    return ''.join(secrets.choice(caracteres) for _ in range(longitud))

print("Tu nueva contraseña:", generar_contrasena(16))
```

creador de codigos qr

Permite generar una imagen QR personalizada a partir de cualquier texto o enlace. Requiere la librería qrcode (ejecuta pip install qrcode[pil]).

```python
import qrcode

def crear_qr(texto, nombre_archivo="codigo.png"):
    img = qrcode.make(texto)
    img.save(nombre_archivo)
    print(f"Código QR guardado como {nombre_archivo}")

crear_qr("https://google.com", "qr_google.png")
```
